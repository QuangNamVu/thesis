```python
import numpy as np
import os
import pandas as pd
home_path = os.path.expanduser("~") + '/'
from utils.load_data import *
from utils.info_params import *
```

```python
data_file_name = home_path + "data/ccxt/preprocessing_data/BTC_USDT_binance_5m.csv"
```

```python
df = pd.read_csv(data_file_name)
df
```

```python
cols = [
       'Open', 'High', 'Low',
       'Close', 'Volume', 'N_buy', 'N_sell', 'buy_amount_avg',
       'sell_amount_avg', 'buy_amount_std', 'sell_amount_std', 'price_avg',
       'price_std', 'cost_avg', 'cost_std', 'Spread_Open_Close',
       'Spread_High_Low', 'MA_Close_6', 'MA_Close_12', 'MA_Close_288',
       'MA_Close_2880', '_1d_Unnamed: 0', '_1d_Timestamp', '_1d_Open',
       '_1d_High', '_1d_Low', '_1d_Close', '_1d_Volume', '_1d_N_buy',
       '_1d_N_sell', '_1d_buy_amount_avg', '_1d_sell_amount_avg',
       '_1d_buy_amount_std', '_1d_sell_amount_std', '_1d_price_avg',
       '_1d_price_std', '_1d_cost_avg', '_1d_cost_std',
       '_1d_Spread_Open_Close', '_1d_Spread_High_Low', '_1d_MA_Close_6',
       '_1d_MA_Close_12', '_1d_MA_Close_288', '_1d_MA_Close_2880',
]
```

```python
X = df[cols].values
y = df['trend'].values
X.shape
N = 25920 # 3 months 3 * 24 * 60 / 5
X_train = X[:-N]
y_train = y[:-N]
X_val = X[-N:]
y_val = y[-N:]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=7 ,random_state=1234, n_estimators=127) #, warm_start = True)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_train_hat = clf.predict(X_train)
cm = confusion_matrix(y_train, y_train_hat)
rp = classification_report(y_train, y_train_hat)
score = accuracy_score(y_train, y_train_hat)

print(cm)
print(rp)
print(score)
```

```python
y_pred = clf.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
rp = classification_report(y_val, y_pred)
score = accuracy_score(y_val, y_pred)

print(cm)
print(rp)
print(score)
```

```python
N = 120
y_pred = clf.predict(X_val[:N])
cm = confusion_matrix(y_val[:N], y_pred)
rp = classification_report(y_val[:N], y_pred)
score = accuracy_score(y_val[:N], y_pred)

print(cm)
print(rp)
print(score)
```

# Save and Load Model:


```python
#Saving model
from sklearn.externals import joblib
joblib.dump(clf, 'checkpoint/Random_Forest_sklearn_5m_3months.joblib')
```
