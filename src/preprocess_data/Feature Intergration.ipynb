```python
import os
import datetime
import pandas as pd
```

```python
home_path = os.path.expanduser("~")

data_file_name = home_path + "/data/ccxt/extra/BTC_USDT_binance_5m.csv"

output_file_name = home_path + "/data/ccxt/preprocessing_data/BTC_USDT_binance_5m.csv"
```

```python
df = pd.read_csv(data_file_name)
df
```

# Spreading Open-Close and High-Low

```python
df['Spread_Open_Close'] = df.Open - df.Close
df['Spread_High_Low'] = df.High - df.Low
```

```python
# columns.append('Spread_Open_Close')
# columns.append('Spread_High_Low')
df['MA_Close_6'] = df['Close'].rolling(6).mean() # 30 mins
df['MA_Close_12'] = df['Close'].rolling(12).mean() # 1 hour
df['MA_Close_288'] = df['Close'].rolling(288).mean() # 1 day
df['MA_Close_2880'] = df['Close'].rolling(2880).mean() # 10 day
```

```python
_1d_cols = df.columns
```

# 1 d diffirent

```python
for att in _1d_cols:
    df['_1d_' + att] = df[att].diff(periods=1)
    
df
```

# Labeling

```python
df['1d_target'] = df.Close.shift(periods=-1) - df.Close
df = df.dropna()
df['trend'] = [1 if x >=0 else 0 for x in df['1d_target']]
df
```

```python
df.to_csv(output_file_name)
print(output_file_name)
```
